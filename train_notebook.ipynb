```python
from ultralytics_custom import YOLO
import class_weight
```

```python
# Load a model

class_weight.apply_weighted_sampling()
data = r"C:\Users\Kevin\Desktop\malaria\codes\Malaria_research\yolov8_baseline\ultralytics\journal-paper-experiment-1-4\data.yaml"
single_cls = True
pretrained = True

model = YOLO(r'C:\Users\Kevin\Desktop\malaria\codes\Malaria_research\yolov8_baseline\ultralytics\Canadia_AI_proposal_Research\configs\YOLO_V8_P0-P1.yaml')  # load yaml and start form scratch
# Train the model
model.train(
    name='yolov8s_1000ep_sgd', 
    data=data,
    pretrained=pretrained,
    epochs = 1,
    imgsz = 640,
    patience = 100,
    batch = 2, #auto batch size
    save_period = -1, #only save final checkpoint and best checkpoint
    optimizer = 'SGD',
    single_cls = single_cls,
    deterministic = False
    )


class_weight.disable_weighted_sampling()

# Evaluate the model
model.val(
    data=data,
    batch=1,
    imgsz=640,
    conf=0.001,  # confidence threshold (default 0.001)
    iou=0.6,  # NMS IOU threshold (default 0.6)
    save_json=True,  # save a cocoapi-compatible JSON results file
    single_cls=single_cls,  # treat as single-class dataset
    deterministic = False
    )
```
